```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random

number_classes = {'Corn': len(os.listdir('./corn weed datasets/corn')),
                 'Bluegrass': len(os.listdir('./corn weed datasets/bluegrass')),
                 'Chenopodium Album': len(os.listdir('./corn weed datasets/chenopodium album')),
                 'Cirsium Setosum': len(os.listdir('./corn weed datasets/cirsium setosum')),
                 'Sedge': len(os.listdir('./corn weed datasets/sedge'))}
```

```python
plt.bar(number_classes.keys(), number_classes.values(), width = .5);
plt.title("Number of Images by Class");
plt.xlabel('Class Name');
plt.ylabel('# of Images');
```

```python
directories = {'Corn': './corn weed datasets/corn',
               'Bluegrass': './corn weed datasets/bluegrass',
               'Chenopodium Album': './corn weed datasets/chenopodium album',
               'Cirsium Setosum': './corn weed datasets/cirsium setosum',
               'Sedge': './corn weed datasets/sedge'}

def get_dims(file):
    '''Returns dimenstions for an RBG image'''
    im = plt.imread(file)
    arr = np.array(im)
    h,w,d = arr.shape
    return h,w
```

```python
for n,d in directories.items():
    lbl = n
    filepath = d
    filelist = [(filepath+'/'+f) for f in os.listdir(filepath)]
    imgs = map(get_dims, filelist)
    
    dim_df = pd.DataFrame(imgs, columns=['height', 'width'])
    sizes = dim_df.groupby(['height', 'width']).size().reset_index().rename(columns={0:'count'})
    sizes.plot.scatter(x='width', y='height')
    plt.title('Image Sizes (pixels) | {}'.format(n))
```

```python
# setting values to rows and column variables
rows = 2
columns = 4
# Adds a subplot
for n, d in directories.items():
    # create figure
    fig = plt.figure(figsize=(10, 7))
    # add images to figure using add_subplot
    for i in range(1,5):
        fig.add_subplot(rows, columns, i)
        img = plt.imread(d+'/'+random.choice(os.listdir(d)))
        plt.imshow(img)
        plt.title('{}{}'.format(n, i))
        plt.axis('off')
```
